**UKURAN PERSEBARAN DATA**<br>
Selain ukuran pemutusan dan  letak  data,  juga  terdapat  ukuran  penyebaran  data.  Ukuran ini berguna untuk menunjukkan seberapa jauhnya suatu data   menyebar   dari  rata-ratanya. Misalnya kita hendak membandingkan   tingkat   jumlah siswa laki-laki dan perempuan yang putus sekolah di Pulau Jawa di Indonesia dalam dua tahun pelajaran. Seumpama  kita  telah  mendapatkan  data bahwa di kedua tahun tersebut memiliki rata-rata jumlah siswa laki-laki dan perempuan yang putus sekolah kurang lebih 1000 siswa pertahun ajaran, di Pulau Jawa Indonesia  namun kita tidak  dapat menyimpulkan bahwa di kedua tahun ajaran tersebut  memiliki tingkat  kenaikan jumlah siswa laki-laki dan perempuan  yang sama karena mungkin saja salah satu dari provinsi cenderung lebih  homogen, dalam  arti  jumlah siswa  laki-laki dan perempuan tidak jauh dari kisaran  rata-rata sedangkan provinsi lainnya  cenderung  heterogen, dalam arti  jumlah  siswa laki-laki dan perempuan jauh  dan sangat  beragam dari kisaran rata-rata.  Untuk itu,  diperlukan  ukuran  penyebaran  data  untuk meneliti  tingkat  jumlah siswa laki-laki dan perempuan yang putus sekolah di Pulau Jawa di Indonesia dalam dua tahun pelajaran. Ukuran  penyebaran  data terdiri dari :
1.   Jangkauan  Data (Range)
2.   Jangkauan  Antar Kuartil
3.   Simpangan  Kuartil
4.   Simpangan  Rata-Rata
5.   Simpangan  Baku (Standar Deviasi)

**A. Jangkauan Data (Range)**<br>
